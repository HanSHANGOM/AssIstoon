```python
import glob
from PIL import Image
import json

def class_crop(classname,savepathbasic,image):
    savepath=savepathbasic+classname+'/'
    clslist=page[classname]
    #print(clslist)
    for clsidx,cls in enumerate(clslist):
        identity=''
        x_min,x_max,y_min,y_max=0,0,0,0
        try:
            x_min=int(cls['@xmin'])
            x_max=int(cls['@xmax'])
            y_min=int(cls['@ymin'])
            y_max=int(cls['@ymax'])
            identity=cls['@id']
            box=(x_min,y_min,x_max,y_max)
            #print(box)
            x_size=x_max-x_min
            y_size=y_max-y_min
        
            size_rate = x_size/y_size
            if size_rate>0.8999 and size_rate<1.1001 and x_size>93 and y_size>93: #100x100 resize를 위한 이미지 비율 범위 선택
                filename =classname+'_'+identity+'.png'
                #+str(x_size)+'x'+str(y_size) #파일명에 사이즈 기재하고 싶을 때 복사 후 삽입
                cropped=image.crop(box) # box 사이즈의 이미지를 crop함
                resized=cropped.resize((100,100))
                resized.save(savepath+filename)
                print(filename+' saved')
        except Exception :
            #print(cls)
            print('err')
        

json_list = glob.glob('X:/Manga109_2017_09_28/annotations_json/*.json')
```
